```python
#importando bibliotecas e classes

import numpy as np  #trabalha com objetos vetor e matriz multidimensional
import matplotlib.pyplot as plt #gera os graficos de visualização
```

```python
# criando os campos descritivos (labels)
labels = ["a1", "a2", "a3" , "a4", "a5"]
```

```python
#criando os valores dos campos (métricas do grafico)
valores = [10, 34, 56, 78, 12]
```

```python
#visualizando o grafico ou objeto plt (matplotlib)
plt.bar(labels,valores)
```

```python
plt.barh(labels,valores)
```

```python
#atribuindo nomes aos eixos do gráfico (x,y)
plt.bar(labels,valores)
plt.xlabel('codigo do produto')
plt.ylabel('valor do produto')
```

```python
#alterar as cores dos elementos graficos
plt.bar(labels,valores,color = "green")
```

```python
# colocando os valores existentes em cada tipo como rotulo de dados

plt.bar(labels,valores,color="green")
plt.xlabel('codigo do produto')
plt.ylabel('valor do produto')

for x,y in zip(labels,valores):
  plt.text(x, y+0.05, '%d' % y, ha='center', va= 'bottom')
```

```python
# criando um novo vetor chamado "valores2"
valores2 = [87, 56, 33, 24, 56] 
```

```python
#plotar os dois elementos juntos
largura = 0.3 
pos1 = np.arange(len(valores))
pos2 = [x + largura for x in pos1]

plt.bar(pos1,valores, width=largura, label='Barra 1', color="red")
plt.bar(pos2, valores2, width=largura, label='Barra 2', color= "black")
plt.legend()
```
